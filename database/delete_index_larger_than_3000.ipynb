# 读取database/dataset中imxx中的编号

```python
import os, re
db_path = "./lab2-image-retrieval/database/dataset"
idx_list = []
for im_file in os.listdir(db_path):
    # 分离数字
    idx = int(re.search(r"\d+", im_file).group())
    idx_list.append(idx)
print(idx_list)
```

# 将tag下所有txt文件中，无法和dataset中imxx对应的编号删去

```python
import os

read_path = "./lab2-image-retrieval/database/tags-incorrect"
save_path = "./lab2-image-retrieval/database/tags"

for file in os.listdir(read_path):
    if file.endswith(".txt"):
        with open(os.path.join(read_path, file), 'r') as f1:
            lines = f1.readlines()
            with open(os.path.join(save_path, file), 'w') as f2:
                for line in lines:
                    line = int(line.strip())
                    if line in idx_list:
                        f2.write(str(line) + '\n')
```
